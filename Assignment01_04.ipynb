```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Month_Wise').getOrCreate()
```

```python
sc=spark.sparkContext
```

```python
rdd1=sc.textFile("/content/sample_data/txn.txt")
```

```python
rdd1.count()
```

```python
for var in rdd1.take(3):
  print(var)
```

```python
rdd2=rdd1.map(lambda a:(a.split(",")[2],a.split(",")[3]))
```

```python
for var in rdd2.collect():
  print(var)
```

```python
kvPairRDD=rdd2.map(lambda a:(a[0],round(float(a[1]),2)))
```

```python
for var in kvPairRDD.take(10):
  print(var)
```

```python
rdd3=kvPairRDD.reduceByKey(lambda a,b:round(a+b,2))
```

```python
for var in rdd3.take(2):
  print(var)
```

```python
sorted=rdd3.sortBy(lambda a: -a[1])
```

```python
for var in sorted.take(10):
  print(var)
```

```python
sorted1=sorted.take(10)
```

```python
rdd4=sc.parallelize(sorted1,10)
```

```python
for var in rdd4.collect():
  print(var)
```

```python
rdd5=sc.textFile("/content/sample_data/custs.txt")
```

```python
for var in rdd5.take(5):
  print(var)
```

```python
rdd6=rdd5.map(lambda a:(a.split(",")[0],a.split(",")[1],a.split(",")[2]))
```

```python
for var in rdd6.take(10):
  print(var)
```

```python
key_c=rdd6.map(lambda a:(a[0],a[1]+" "+a[2]))
```

```python
for i in key_c.take(5):
  print(i)
```

```python
joinedRDD=key_c.join(rdd4)
```

```python
for i in joinedRDD.take(3):
  print(i)
```

```python
rdd7=joinedRDD.map(lambda a:(a[1][0],round(a[1][1],2)))
```

```python
for i in rdd7.take(3):
  print(i)
```

```python
sorted_final=rdd7.sortBy(lambda a:-(a[1]))
```

```python
for i in sorted_final.collect():
  print(i)
```

```python
#reformat
csv=sorted_final.map(lambda a:a[0]+","+str(a[1]))
```

```python
csv.getNumPartitions()
```

```python
csv=csv.coalesce(1)
```

```python
csv_output=csv.saveAsTextFile("/content/sample_data/Top10Cust")
```
