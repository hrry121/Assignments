```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Month_Wise').getOrCreate()
```

```python
sc=spark.sparkContext
```

```python
rdd1=sc.textFile("/content/sample_data/txn.txt")
```

```python
rdd1.getNumPartitions()
```

```python
rdd1.count()
```

```python
import datetime
from datetime import datetime
```

```python
format_date="%m-%d-%Y"
```

```python
tupleRDD = rdd1.map(lambda a :(datetime.strptime(a.split(",")[1], format_date).strftime("%-m"),
                               datetime.strptime(a.split(",")[1], format_date).strftime("%B"),datetime.strptime(a.split(",")[1], format_date).strftime("%Y"),
                              a.split(",")[3]))
```

```python
for var in tupleRDD.collect():
  print(var)
```

```python
kvPairRDD=tupleRDD.map(lambda a:((a[0])+" "+a[1]+" "+(a[2]),float(a[3])))
```

```python
for var in kvPairRDD.take(3):
  print(var)
```

```python
total=kvPairRDD.reduceByKey(lambda a,b:a+b)
```

```python
for var in total.take(3):
  print(var)
```

```python
sortedRDD = total.sortBy(lambda a: int(a[0].split()[0]))
```

```python
for var in sortedRDD.collect():
  print(var)
```

```python
sortedRDD.getNumPartitions()
```

```python
sortedRDD=sortedRDD.coalesce(1)
```

```python
#Format the answer
csv_formatRDD=sortedRDD.map(lambda a: a[0].split()[1]+","+a[0].split()[2]+","+str(round(a[1],2)))
```

```python
for var in csv_formatRDD.collect():
  print(var)
```

```python
csv_formatRDD.saveAsTextFile("/content/sample_data/Month_Wise_Sales1")
```
